#IMPORTING LIBRARIES

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pt
%matplotlib inline
```

#IMPORTING THE DATASET

```python
iris=datasets.load_iris()
iris
```

#CREATING THE DATAFRAME

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
```

#CREATING THE TARGET VARIABLE

```python
y=iris.target
df['target']=y
df.head()
```

#CHECK THE MISSING VALUES

```python
df.apply(lambda x:sum(x.isnull()))
```

#SPLITTING THE DATA AND DEFINING THE DECISION TREE ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#Splitting the data into training and testing sets
X=df.drop('target',axis=1)
y=df['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

#defining(training) the decision tree
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

#VISUALIZING THE DECISION TREE

```python
#set the predictions
prediction=dtree.predict(X_test)
```

```python
from sklearn import tree
names=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
classes=['setosa','versicolor','virginica']
fig,axes=pt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(dtree,feature_names=names,class_names=classes,filled=True);
```
